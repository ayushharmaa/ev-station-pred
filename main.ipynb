```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns

# Step 1: Select Features for Clustering
X_cluster = data[['Population', 'Income', 'Vehicles']]

# Step 2: Normalize the Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

# Step 3: Find the Optimal Number of Clusters (Elbow Method)
inertia = []
for k in range(1, 11):  # Try clusters from 1 to 10
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title("Elbow Method for Optimal Clusters", fontsize=14)
plt.xlabel("Number of Clusters", fontsize=12)
plt.ylabel("Inertia", fontsize=12)
plt.grid(True)
plt.show()

# Step 4: Apply K-Means Clustering (Using Optimal Clusters)
optimal_clusters = 3  # Replace this with the value chosen from the Elbow Method
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
data['Cluster'] = kmeans.fit_predict(X_scaled)

# Step 5: Visualize the Clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(data=data, x='Population', y='Vehicles', hue='Cluster', palette='Set1', s=100)
plt.title("K-Means Clustering: Population vs Vehicles", fontsize=14)
plt.xlabel("Population", fontsize=12)
plt.ylabel("Vehicles", fontsize=12)
plt.legend(title="Cluster")
plt.grid(True)
plt.show()

# Optional: 3D Visualization (if you want to include Income)
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(data['Population'], data['Vehicles'], data['Income'], c=data['Cluster'], cmap='Set1', s=50)
ax.set_title("3D Visualization of Clusters")
ax.set_xlabel("Population")
ax.set_ylabel("Vehicles")
ax.set_zlabel("Income")
plt.colorbar(scatter)
plt.show()

# Print Cluster Assignments
print("\nCluster Assignments for Each Ward:")
print(data[['Ward', 'Cluster']])

# Assign weights to clusters based on priority
cluster_priority = {
    0: (0, 25),    # Cluster 0: Low priority
    1: (25, 50),   # Cluster 1: Medium priority
    2: (50, 100)   # Cluster 2: High priority
}

# Create a function to assign probabilities
def calculate_success_probability(row):
    # Get base range from cluster priority
    cluster_range = cluster_priority[row['Cluster']]

    # Scale within the range based on vehicle count
    scaled_probability = (row['Vehicles'] - data['Vehicles'].min()) / (data['Vehicles'].max() - data['Vehicles'].min())
    return cluster_range[0] + scaled_probability * (cluster_range[1] - cluster_range[0])

# Apply the function to calculate success probabilities
data['Success_Probability'] = data.apply(calculate_success_probability, axis=1)

# Assign success categories based on probabilities
def assign_success_category(prob):
    if prob <= 25:
        return "Low"
    elif 25 < prob <= 50:
        return "Medium"
    elif 50 < prob <= 100:
        return "High"

data['Success_Category'] = data['Success_Probability'].apply(assign_success_category)

# Display the results
pd.set_option('display.max_rows', None)
print(data[['Ward', 'Cluster', 'Vehicles', 'Success_Probability', 'Success_Category']])

```
